```python
from bs4 import BeautifulSoup # 웹에서 가져온 데이터를 처리
import pandas as pd           # 데이터 분석 필수
```

```python
driver = webdriver.Chrome('chromedriver.exe') #파이썬으로 크롬 브라우저 제어 크롬 버전은 동일버전으로 다운
```

```python
url = "https://www.starbucks.co.kr/store/store_map.do?disp=locale"
driver.get(url)  # 특정 웹페이지로 크롬을 통해 이동
```

```python
seoul_btn = '#container > div > form > fieldset > div > section > article.find_store_cont > article > article:nth-child(4) > div.loca_step1 > div.loca_step1_cont > ul > li:nth-child(1) > a'
driver.find_element_by_css_selector(seoul_btn).click()  

# 화면 상의 특정 위치를 선택하여 클릭
```

```python
all_btn = '#mCSB_2_container > ul > li:nth-child(1) > a'
driver.find_element_by_css_selector(all_btn).click()  
```

```python
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
```

```python
print(soup)
```

```python
starbucks_soup_list = soup.select('li.quickResultLstCon')
print(len(starbucks_soup_list))
```

```python
print(starbucks_soup_list)
```

```python
starbucks_store = starbucks_soup_list[0]
name = starbucks_store.select('strong')[0].text.strip()  # html tag
lat = starbucks_store['data-lat'].strip()   # 위치정보 위도
lng = starbucks_store['data-long'].strip()  # 위치정보 경도
address = str(starbucks_store.select('p.result_details')[0]).split('<br/>')[0].split('>')[1]  # 매장 주소
tel = str(starbucks_store.select('p.result_details')[0]).split('<br/>')[1].split('<')[0]   # 매장 전화번호


print(name) 
print(lat) 
print(lng)
print(address)
print(tel)
```

```python
starbucks_list = []
for item in starbucks_soup_list:
    
    name = item.select('strong')[0].text.strip()  # html tag
    lat = item['data-lat'].strip()   # 위치정보 위도
    lng = item['data-long'].strip()  # 위치정보 경도
    address = str(item.select('p.result_details')[0]).split('<br/>')[0].split('>')[1]  # 매장 주소
    tel = str(item.select('p.result_details')[0]).split('<br/>')[1].split('<')[0]   # 매장 전화번호
    
    starbucks_list.append([ name, lat, lng, address, tel])    
```

```python
print(starbucks_list)
```
